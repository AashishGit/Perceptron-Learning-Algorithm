## Perceptron Learning Algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#Load all the datasets
inputData = pd.read_csv('inputData.csv', header = None, index_col = None)
inputData20 = pd.read_csv('inputData20.csv', header = None, index_col = None)
inputData100 = pd.read_csv('inputData100.csv', header = None, index_col = None)
inputData1000 = pd.read_csv('inputData1000.csv', header = None, index_col = None)
```

```python
def create_dataset(input_data, w):
    ##Evaluate yn for each xn{x1,x2} using the target fn
    ## f = sign(1+5*x1 + 2*x2)
    xn = zip(input_data[0], input_data[1])
    x1, x2 = zip(*xn)
    x1 = list(x1)
    x2 = list(x2)
    yn = []
    for i in range(len(x1)):
        f = w[0] + w[1]*x1[i] + w[2]*x2[i]
        if f > 0:
            fn = 1

        elif f < 0:
            fn = -1

        yn.append(fn)

    Dataset = []
    for i in range(len(yn)):
        Dataset.append([x1[i], x2[i], yn[i]])
        
    inputData = pd.DataFrame()
    inputData[0] = x1
    inputData[1] = x2
    inputData[2] = yn
        
    return Dataset, inputData
```

```python
def Perceptron(n_iter, w, data):
    for iteration in range(n_iter):
        squared_error = 0.0
        for row in data:
            h = np.dot(row[:2], w[1:]) + w[0]
            if h>0:
                y_pred = 1
            else:
                y_pred = -1
            error = row[-1] - y_pred
            squared_error += error**2
            w[0] += error
            for i in range(len(row)-1):
                w[i+1] += error*row[i]
        print('>iteration=%d, >error=%.3f' % (iteration, squared_error))
    updated_w = w
        
    return updated_w
```

```python
#Create labelled data for all the datasets
data, data_plotting = create_dataset(inputData, [1, 5, 2])
data20, data_plotting20 = create_dataset(inputData20, [1, 5, 2])
data100, data_plotting100 = create_dataset(inputData100, [1, 5, 2])
data1000, data_plotting1000 = create_dataset(inputData1000, [1, 5, 2])
```

```python
#No. of iterations
n_iter = 25
#initialize weights for each dataset
w = [0.0 for i in range(len(data[0]))]
w20 = [0.0 for i in range(len(data20[0]))]
w100 = [0.0 for i in range(len(data100[0]))]
w1000 = [0.0 for i in range(len(data1000[0]))]
```

```python
#For inputData the PLA converges in 1 iteration
perceptron_weights = Perceptron(n_iter, w, data)
```

```python
#For inputData20 the PLA converges in 1 iteration
perceptron_weights20 = Perceptron(n_iter, w20, data20)
```

```python
#For inputData100 the PLA converges in 16 iterations
perceptron_weights100 = Perceptron(n_iter, w100, data100)
```

```python
#For inoutData1000 the PLA converges in 23 iterations
perceptron_weights1000 = Perceptron(n_iter, w1000, data1000)
```

## Plot outputs and interprete results

```python
def Set_Colors_Array(true_labels):
    col = [None]*len(true_labels)
    for i in range(len(col)):
        if true_labels[i] == -1:
            col[i] = 'r'
        else:
            col[i] = 'g'
    return col
```

```python
def plot_point_and_target(data):
    colors = Set_Colors_Array(data[2])
    x_1 = data[0]
    x_2 = data[1]
    y = (-1-5*x_1)/2
    plt.plot(x_1,y)
    plt.scatter(x_1,x_2,c=colors)
    plt.xlabel('x1')
    plt.ylabel('x2')
#     plt.imsave('point_and_oracle_fn.jpg')
    plt.show()
```

```python
plot_point_and_target(data_plotting), plot_point_and_target(data_plotting20)
```

```python
plot_point_and_target(data_plotting100), plot_point_and_target(data_plotting1000)
```

```python
def plot_point_oracle_and_hypothesis(data):
    colors = Set_Colors_Array(data[2])
    x_1 = data[0]
    y = (-1-5*x_1)/2
    z = (-w[0]-w[1]*x_1)/w[2]
    plt.plot(x_1,y)
    plt.plot(x_1,z)
    plt.legend(['f(x)','g(x)'])
    x_2 = data[1]
    plt.scatter(x_1,x_2,c = colors)
    plt.show()
```

```python
plot_point_oracle_and_hypothesis(data_plotting), plot_point_oracle_and_hypothesis(data_plotting20)
```

- For inputData f is very close to g as all the data points are correctly fitted by the hypothesis line
- For inputData20 the distance between f and g is very small as almost all the data points are fitted correctly by the hypothesis line

```python
plot_point_oracle_and_hypothesis(data_plotting100), plot_point_oracle_and_hypothesis(data_plotting1000)
```

- As the number of data points increases and the space becomes more populated the hypothesis starts making some error sover some finite number of datapoints, but still doesn;t deviate from f and stays close to f
